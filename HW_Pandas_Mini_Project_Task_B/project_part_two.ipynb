```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
file_df = pd.read_csv(file)
file_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_file_df = file_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_file_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
reduced_file_df= reduced_file_df.replace({0.0: "No", 1.0: "yes"})
reduced_file_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = reduced_file_df.loc[reduced_file_df["AttendedBootcamp"] == "yes"]
attended_bootcamp.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
# Create DataFrame of the different boot camps that had a significant number of attendees
Bootcamp_Name = pd.DataFrame(reduced_file_df["BootcampName"].value_counts())

Bootcamp_Name.reset_index(inplace=True)
Bootcamp_Name.columns = ["BootcampName", "Count"]

Bootcamp_Name = Bootcamp_Name.loc[(Bootcamp_Name["Count"] > 9) & (Bootcamp_Name["BootcampName"] != 'Free Code Camp is not a bootcamp - please scroll up and charge answer to "no"')]
Bootcamp_Name.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
attended_bootcamp = reduced_file_df.loc[reduced_file_df["BootcampRecommend"] == "yes"]
attended_bootcamp.count()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
Recommend_bootcamp = pd.DataFrame(reduced_file_df["BootcampRecommend"].value_counts())

Recommend_bootcamp.reset_index(inplace=True)
Recommend_bootcamp.columns = ["BootcampRecommend", "Count"]

Recommend_bootcamp = Recommend_bootcamp.loc[(Recommend_bootcamp["Count"] > 9) & (Recommend_bootcamp["BootcampRecommend"] != 'Free Code Camp is not a bootcamp - please scroll up and charge answer to "no"')]
Recommend_bootcamp.head()
```

```python
Recommend_bootcamp = attended_bootcamp.replace({"Yes": 1, "No": 0})
Recommend_bootcamp = pd.DataFrame(Recommend_bootcamp.groupby("BootcampName")["BootcampRecommend"].count())

Recommend_bootcamp.reset_index(inplace=True)
Recommend_bootcamp.columns=["BootcampName", "Recommenders"]

Recommend_bootcamp.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
BootcampRecommend = Recommend_bootcamp["BootcampName"].value_counts()
BootcampRecommend.head(25)
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged_df = pd.merge(Recommend_bootcamp, Bootcamp_Name, on="BootcampName")
merged_df.head()
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
percent_Recommenders = (Recommend_bootcamp['Recommenders'] /Bootcamp_Name['Count'] *100).astype(float).map(
    "{:,.2f}%".format)
percent_Recommenders.head()
```
